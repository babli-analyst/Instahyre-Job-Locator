```python
df = pd.read_csv('scrapped_886.csv')
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df['job_title'] = df['job_title'].str.replace("(Drive)","")
df['founded_in'] = df['founded_in'].str.replace("Founded in ","")
df['employees_count'] = df['employees_count'].str.replace("More than ","").str.replace(" employees","")
```

```python
df
```

```python
df['job_title'] = df['job_title'].str.replace(")","").str.replace("(","")
```

```python
df
```

```python
df[["e1","e2"]] = df['employees_count'].str.split(" - ",expand = True)
```

```python
df
```

```python
import numpy as np
```

```python
df = df.drop("employees_count",axis=1)
df['e1'] = df['e1'].replace("None",np.nan).astype(float)
df['e2'] = df['e2'].replace("None",np.nan).astype(float)
```

```python
df['avg_employees_count'] = df[["e1","e2"]].mean(axis=1)
```

```python
df = df.drop(["e1","e2"],axis=1)
```
